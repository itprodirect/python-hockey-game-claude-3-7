```python
import pygame
import sys
import random
import time

# Initialize pygame
pygame.init()

# Screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Tampa Bay Lightning Shootout Challenge")

# Colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
BLUE = (0, 119, 192)  # Tampa Bay Lightning blue
LIGHT_BLUE = (150, 200, 255)
RED = (255, 0, 0)
ICE_COLOR = (240, 248, 255)

# Game variables
clock = pygame.time.Clock()
FPS = 60
level = 1
max_level = 5
score = 0
attempts = 0
max_attempts = 5
game_active = True
game_won = False
game_over = False

# Font
font = pygame.font.SysFont(None, 36)
title_font = pygame.font.SysFont(None, 48)

# Player (shooter) properties
player_width, player_height = 30, 50
player_x = WIDTH // 4
player_y = HEIGHT // 2 - player_height // 2
player_speed = 5

# Goalie properties
goalie_width, goalie_height = 40, 60
goalie_x = WIDTH - 100
goalie_y = HEIGHT // 2 - goalie_height // 2
goalie_speed = 4  # Base speed, will increase with levels

# Puck properties
puck_radius = 10
puck_x = player_x + player_width + 5
puck_y = player_y + player_height // 2
puck_speed = 6  # Base speed, will increase with levels
puck_active = False

# Function to reset puck position
def reset_puck():
    global puck_x, puck_y, puck_active
    puck_x = player_x + player_width + 5
    puck_y = player_y + player_height // 2
    puck_active = False

# Function to increase difficulty based on level
def set_level_difficulty(level):
    global goalie_speed, puck_speed
    goalie_speed = 4 + level * 0.8
    puck_speed = 6 + level * 1.2

# Game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            # Shoot the puck
            if event.key == pygame.K_SPACE and not puck_active and game_active:
                puck_active = True
            # Restart game if over
            if (event.key == pygame.K_r) and (game_over or game_won):
                # Reset game
                level = 1
                score = 0
                attempts = 0
                game_active = True
                game_over = False
                game_won = False
                set_level_difficulty(level)
                reset_puck()
    
    # Fill background
    screen.fill(ICE_COLOR)
    
    # Draw rink markings
    pygame.draw.rect(screen, BLUE, (0, 0, WIDTH, HEIGHT), 10)  # Border
    pygame.draw.circle(screen, RED, (WIDTH // 2, HEIGHT // 2), 50, 2)  # Center circle
    pygame.draw.line(screen, RED, (WIDTH // 2, 0), (WIDTH // 2, HEIGHT), 2)  # Center line
    
    # Draw the goal
    goal_width, goal_height = 80, 120
    goal_x = WIDTH - 50
    goal_y = HEIGHT // 2 - goal_height // 2
    pygame.draw.rect(screen, RED, (goal_x, goal_y, goal_width, goal_height), 3)
    
    if game_active:
        # Player movement
        keys = pygame.key.get_pressed()
        if keys[pygame.K_UP] and player_y > 10:
            player_y -= player_speed
        if keys[pygame.K_DOWN] and player_y < HEIGHT - player_height - 10:
            player_y += player_speed
            
        # If puck not active, it follows the player
        if not puck_active:
            puck_y = player_y + player_height // 2
            
        # AI Goalie movement
        # Goalie follows the puck's y position if active, with some lag
        target_y = puck_y - goalie_height // 2
        if goalie_y < target_y:
            goalie_y += min(goalie_speed, target_y - goalie_y)
        elif goalie_y > target_y:
            goalie_y -= min(goalie_speed, goalie_y - target_y)
        
        # Ensure goalie stays within the rink
        goalie_y = max(10, min(HEIGHT - goalie_height - 10, goalie_y))
        
        # Puck movement when active
        if puck_active:
            puck_x += puck_speed
            
            # Check for goal
            if puck_x >= goal_x and goal_y <= puck_y <= goal_y + goal_height:
                score += 1
                attempts += 1
                puck_active = False
                reset_puck()
                
                # Level up if scored enough
                if score >= level and level < max_level:
                    level += 1
                    set_level_difficulty(level)
                
                # Check for win
                if level == max_level and score >= max_level:
                    game_active = False
                    game_won = True
            
            # Check for goalie save
            if (goalie_x <= puck_x <= goalie_x + goalie_width and 
                goalie_y <= puck_y <= goalie_y + goalie_height):
                attempts += 1
                puck_active = False
                reset_puck()
            
            # Check if puck is out of bounds
            if puck_x > WIDTH:
                attempts += 1
                puck_active = False
                reset_puck()
        
        # Check for game over
        if attempts >= max_attempts and not game_won:
            game_active = False
            game_over = True
        
        # Draw player (shooter)
        pygame.draw.rect(screen, BLUE, (player_x, player_y, player_width, player_height))
        
        # Draw goalie
        pygame.draw.rect(screen, WHITE, (goalie_x, goalie_y, goalie_width, goalie_height))
        
        # Draw puck
        if puck_active or not game_active:
            pygame.draw.circle(screen, BLACK, (int(puck_x), int(puck_y)), puck_radius)
        
        # Display score and level
        score_text = font.render(f"Score: {score}/{level}", True, BLACK)
        screen.blit(score_text, (20, 20))
        
        level_text = font.render(f"Level: {level}/{max_level}", True, BLACK)
        screen.blit(level_text, (20, 60))
        
        attempts_text = font.render(f"Attempts: {attempts}/{max_attempts}", True, BLACK)
        screen.blit(attempts_text, (20, 100))
        
        # Display controls
        controls_text = font.render("UP/DOWN: Move   SPACE: Shoot", True, BLACK)
        screen.blit(controls_text, (WIDTH // 2 - 180, HEIGHT - 40))
    
    # Game over screens
    if game_over:
        overlay = pygame.Surface((WIDTH, HEIGHT), pygame.SRCALPHA)
        overlay.fill((0, 0, 0, 128))
        screen.blit(overlay, (0, 0))
        
        game_over_text = title_font.render("GAME OVER", True, RED)
        screen.blit(game_over_text, (WIDTH // 2 - game_over_text.get_width() // 2, HEIGHT // 2 - 50))
        
        restart_text = font.render("Press R to Restart", True, WHITE)
        screen.blit(restart_text, (WIDTH // 2 - restart_text.get_width() // 2, HEIGHT // 2 + 20))
        
        final_score = font.render(f"Final Score: {score}", True, WHITE)
        screen.blit(final_score, (WIDTH // 2 - final_score.get_width() // 2, HEIGHT // 2 + 70))
    
    if game_won:
        overlay = pygame.Surface((WIDTH, HEIGHT), pygame.SRCALPHA)
        overlay.fill((0, 0, 0, 128))
        screen.blit(overlay, (0, 0))
        
        win_text = title_font.render("YOU WIN!", True, LIGHT_BLUE)
        screen.blit(win_text, (WIDTH // 2 - win_text.get_width() // 2, HEIGHT // 2 - 50))
        
        champion_text = font.render("Lightning Cup Champion!", True, WHITE)
        screen.blit(champion_text, (WIDTH // 2 - champion_text.get_width() // 2, HEIGHT // 2 + 20))
        
        restart_text = font.render("Press R to Play Again", True, WHITE)
        screen.blit(restart_text, (WIDTH // 2 - restart_text.get_width() // 2, HEIGHT // 2 + 70))
    
    # Update the display
    pygame.display.flip()
    
    # Cap the frame rate
    clock.tick(FPS)

# Quit the game
pygame.quit()
sys.exit()
```

```python
import pygame
import sys
import random
import time
import os
import math

# Initialize pygame
pygame.init()

# Screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Tampa Bay Lightning Shootout Challenge")

# Colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
BLUE = (0, 119, 192)       # Tampa Bay Lightning blue
LIGHT_BLUE = (150, 200, 255)
LIGHTNING_BLUE = (0, 36, 91)  # Tampa Bay Lightning dark blue
LIGHTNING_SILVER = (163, 169, 174)  # Tampa Bay Lightning silver
RED = (255, 0, 0)
DARK_GRAY = (30, 30, 40)   # Dark mode background
ICE_COLOR = (200, 220, 235) # Slightly blue-tinted ice for dark mode

# Game variables
clock = pygame.time.Clock()
FPS = 60
level = 1
max_level = 5
score = 0
attempts = 0
max_attempts = 10  # Increased from 5 to 10 shots
game_active = True
game_won = False
game_over = False
celebration_active = False
celebration_timer = 0
celebration_duration = 2 * FPS  # 2 seconds worth of frames

# Font
font = pygame.font.SysFont(None, 36)
title_font = pygame.font.SysFont(None, 48)

# Player (shooter) properties
player_width, player_height = 40, 60
player_x = WIDTH // 4
player_y = HEIGHT // 2 - player_height // 2
player_speed = 7

# Goalie properties
goalie_width, goalie_height = 50, 70
goalie_x = WIDTH - 100
goalie_y = HEIGHT // 2 - goalie_height // 2
base_goalie_speed = 3

# Puck properties
puck_radius = 10
puck_x = player_x + player_width + 5
puck_y = player_y + player_height // 2
base_puck_speed = 8
puck_active = False

# Goal dimensions
goal_width, goal_height = 80, 120

# Create placeholder player images
def create_player_image():
    surface = pygame.Surface((player_width, player_height), pygame.SRCALPHA)
    # Player body (blue jersey)
    pygame.draw.rect(surface, LIGHTNING_BLUE, (0, 10, player_width, player_height-20))
    # Player head
    pygame.draw.circle(surface, (255, 213, 170), (player_width//2, 10), 10)
    # Player helmet
    pygame.draw.circle(surface, BLUE, (player_width//2, 8), 10, 2)
    # Jersey number
    number_font = pygame.font.SysFont(None, 20)
    number = number_font.render("91", True, WHITE)  # Stamkos #91
    surface.blit(number, (player_width//2 - 6, player_height//2 - 5))
    # Stick
    pygame.draw.line(surface, BLACK, (player_width-5, player_height-15), 
                    (player_width+15, player_height-5), 2)
    return surface

# Create goalie image
def create_goalie_image():
    surface = pygame.Surface((goalie_width, goalie_height), pygame.SRCALPHA)
    # Goalie pads (white)
    pygame.draw.rect(surface, WHITE, (0, 0, goalie_width, goalie_height))
    # Goalie jersey (blue)
    pygame.draw.rect(surface, LIGHTNING_BLUE, (5, 5, goalie_width-10, goalie_height//2))
    # Goalie mask
    pygame.draw.circle(surface, WHITE, (goalie_width//2, goalie_height//4), 10)
    # Goalie mask details
    pygame.draw.circle(surface, LIGHTNING_BLUE, (goalie_width//2, goalie_height//4), 10, 2)
    # Blocker and glove
    pygame.draw.rect(surface, LIGHTNING_SILVER, (0, goalie_height//2+5, 8, 15))
    pygame.draw.circle(surface, LIGHTNING_SILVER, (goalie_width-5, goalie_height//2+10), 8)
    return surface

# Create Tampa Bay Lightning logo
def create_lightning_logo():
    surface = pygame.Surface((100, 100), pygame.SRCALPHA)
    # Circle background
    pygame.draw.circle(surface, LIGHTNING_BLUE, (50, 50), 45)
    pygame.draw.circle(surface, WHITE, (50, 50), 45, 2)
    
    # Lightning bolt (simplified)
    points = [
        (40, 20),
        (55, 35),
        (45, 50),
        (60, 65),
        (35, 80),
        (45, 55),
        (30, 40)
    ]
    pygame.draw.polygon(surface, WHITE, points)
    return surface

# Create goal image
def create_goal_image():
    surface = pygame.Surface((goal_width+10, goal_height+10), pygame.SRCALPHA)
    # Net frame
    pygame.draw.rect(surface, RED, (0, 0, goal_width, goal_height), 3)
    # Net lines - vertical
    for i in range(1, 8):
        x = i * (goal_width // 8)
        pygame.draw.line(surface, (150, 150, 150), (x, 0), (x, goal_height), 1)
    # Net lines - horizontal
    for i in range(1, 6):
        y = i * (goal_height // 6)
        pygame.draw.line(surface, (150, 150, 150), (0, y), (goal_width, y), 1)
    return surface

# Create puck image
def create_puck_image():
    surface = pygame.Surface((puck_radius*2+2, puck_radius*2+2), pygame.SRCALPHA)
    pygame.draw.circle(surface, BLACK, (puck_radius+1, puck_radius+1), puck_radius)
    pygame.draw.circle(surface, LIGHTNING_SILVER, (puck_radius+1, puck_radius+1), puck_radius, 1)
    return surface

# Create celebration effects
def create_celebration_effect(frame):
    effects = []
    
    # Strobe lights - alternating colors every few frames
    if frame % 8 < 4:
        color = BLUE
    else:
        color = WHITE
    
    # Add lightning bolts that move/change with each frame
    for i in range(5):
        angle = (frame * 5 + i * 72) % 360
        distance = 150 + 20 * math.sin(frame * 0.1 + i)
        x = WIDTH // 2 + distance * math.cos(math.radians(angle))
        y = HEIGHT // 2 + distance * math.sin(math.radians(angle))
        
        bolt = {
            'type': 'lightning',
            'x': x,
            'y': y,
            'size': 30 + 10 * math.sin(frame * 0.2 + i),
            'color': color
        }
        effects.append(bolt)
    
    # Add celebratory text that pulses
    text_scale = 1.0 + 0.2 * math.sin(frame * 0.1)
    text = {
        'type': 'text',
        'text': "GOAL!",
        'size': int(60 * text_scale),
        'color': WHITE if frame % 10 < 5 else BLUE,
        'x': WIDTH // 2,
        'y': HEIGHT // 2 - 50
    }
    effects.append(text)
    
    return effects

# Create images
player_img = create_player_image()
goalie_img = create_goalie_image()
lightning_logo = create_lightning_logo()
goal_img = create_goal_image()
puck_img = create_puck_image()

# Function to draw a lightning bolt
def draw_lightning_bolt(surface, x, y, size, color=LIGHT_BLUE):
    points = [
        (x, y - size//2),
        (x + size//4, y - size//4),
        (x - size//8, y + size//8),
        (x + size//8, y + size//4),
        (x - size//4, y + size//2),
        (x + size//6, y + size//6),
        (x - size//6, y)
    ]
    pygame.draw.polygon(surface, color, points)
    pygame.draw.polygon(surface, WHITE, points, 2)

# Function to reset puck position
def reset_puck():
    global puck_x, puck_y, puck_active
    puck_x = player_x + player_width + 5
    puck_y = player_y + player_height // 2
    puck_active = False

# Function to increase difficulty based on level
def set_level_difficulty(level):
    global goalie_speed, puck_speed
    # Make early levels easier, later levels harder
    goalie_speed = base_goalie_speed + (level - 1) * 0.8
    puck_speed = base_puck_speed + (level - 1) * 1.0
    
    # First level is much easier
    if level == 1:
        goalie_speed = base_goalie_speed * 0.6
    
    # Final level is extra challenging
    if level == max_level:
        goalie_speed = base_goalie_speed + 4.0

# Set initial difficulty
set_level_difficulty(level)

# Game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            # Shoot the puck
            if event.key == pygame.K_SPACE and not puck_active and game_active and not celebration_active:
                puck_active = True
            # Restart game if over
            if (event.key == pygame.K_r) and (game_over or game_won):
                # Reset game
                level = 1
                score = 0
                attempts = 0
                game_active = True
                game_over = False
                game_won = False
                celebration_active = False
                set_level_difficulty(level)
                reset_puck()
            # Skip celebration
            if event.key == pygame.K_SPACE and celebration_active:
                celebration_active = False
                celebration_timer = 0
    
    # Fill background with dark color
    screen.fill(DARK_GRAY)
    
    # Draw ice rink
    pygame.draw.rect(screen, ICE_COLOR, (20, 20, WIDTH-40, HEIGHT-40))
    
    # Draw rink markings
    pygame.draw.rect(screen, LIGHTNING_BLUE, (20, 20, WIDTH-40, HEIGHT-40), 10)  # Border
    pygame.draw.circle(screen, RED, (WIDTH // 2, HEIGHT // 2), 50, 2)  # Center circle
    pygame.draw.line(screen, RED, (WIDTH // 2, 20), (WIDTH // 2, HEIGHT-20), 2)  # Center line
    
    # Draw Lightning logo at center ice
    screen.blit(lightning_logo, (WIDTH // 2 - 50, HEIGHT // 2 - 50))
    
    # Draw the goal
    goal_x = WIDTH - 50
    goal_y = HEIGHT // 2 - goal_height // 2
    screen.blit(goal_img, (goal_x-5, goal_y-5))
    
    # Draw level indicator in a Tampa Bay style box
    level_box = pygame.Rect(WIDTH-200, 20, 180, 80)
    pygame.draw.rect(screen, LIGHTNING_BLUE, level_box)
    pygame.draw.rect(screen, LIGHTNING_SILVER, level_box, 2)
    
    level_text = font.render(f"LEVEL: {level}/{max_level}", True, WHITE)
    screen.blit(level_text, (WIDTH-180, 30))
    
    score_text = font.render(f"SCORE: {score}/{level}", True, WHITE)
    screen.blit(score_text, (WIDTH-180, 60))
    
    # Attempts shown separately
    attempts_box = pygame.Rect(20, 20, 180, 40)
    pygame.draw.rect(screen, LIGHTNING_BLUE, attempts_box)
    pygame.draw.rect(screen, LIGHTNING_SILVER, attempts_box, 2)
    
    attempts_text = font.render(f"SHOTS: {attempts}/{max_attempts}", True, WHITE)
    screen.blit(attempts_text, (30, 30))
    
    # Celebration handling
    if celebration_active:
        celebration_timer += 1
        if celebration_timer >= celebration_duration:
            celebration_active = False
            celebration_timer = 0
        else:
            # Draw strobe and celebration effects
            effects = create_celebration_effect(celebration_timer)
            
            # Render all effects
            for effect in effects:
                if effect['type'] == 'lightning':
                    draw_lightning_bolt(screen, effect['x'], effect['y'], effect['size'], effect['color'])
                elif effect['type'] == 'text':
                    text_surf = pygame.font.SysFont(None, effect['size']).render(
                        effect['text'], True, effect['color'])
                    screen.blit(text_surf, 
                               (effect['x'] - text_surf.get_width()//2, 
                                effect['y'] - text_surf.get_height()//2))
            
            # Continue showing the game in the background
    
    if game_active and not celebration_active:
        # Player movement
        keys = pygame.key.get_pressed()
        if keys[pygame.K_UP] and player_y > 30:
            player_y -= player_speed
        if keys[pygame.K_DOWN] and player_y < HEIGHT - player_height - 30:
            player_y += player_speed
            
        # If puck not active, it follows the player
        if not puck_active:
            puck_y = player_y + player_height // 2
            
        # AI Goalie movement
        # Make goalie behavior more predictable in early levels
        if level <= 2:
            # In early levels, goalie has a delay following the puck
            target_y = puck_y - goalie_height // 2
            if random.random() < 0.15:  # Occasionally make a mistake
                target_y += random.randint(-40, 40)
        else:
            # In later levels, goalie follows more accurately
            target_y = puck_y - goalie_height // 2
            if random.random() < 0.05:  # Less frequent mistakes
                target_y += random.randint(-20, 20)
        
        # Move goalie toward target
        if goalie_y < target_y:
            goalie_y += min(goalie_speed, target_y - goalie_y)
        elif goalie_y > target_y:
            goalie_y -= min(goalie_speed, goalie_y - target_y)
        
        # Ensure goalie stays within the rink
        goalie_y = max(30, min(HEIGHT - goalie_height - 30, goalie_y))
        
        # Puck movement when active
        if puck_active:
            puck_x += puck_speed
            
            # Check for goal - expanded goal hitbox slightly to make scoring easier
            extended_goal_y = goal_y - 10
            extended_goal_height = goal_height + 20
            if puck_x >= goal_x and extended_goal_y <= puck_y <= extended_goal_y + extended_goal_height:
                score += 1
                attempts += 1
                puck_active = False
                
                # Start celebration
                celebration_active = True
                celebration_timer = 0
                
                reset_puck()
                
                # Level up if scored enough
                if score >= level and level < max_level:
                    level += 1
                    set_level_difficulty(level)
                
                # Check for win
                if level == max_level and score >= max_level:
                    game_active = False
                    game_won = True
            
            # Check for goalie save
            if (goalie_x <= puck_x <= goalie_x + goalie_width and 
                goalie_y <= puck_y <= goalie_y + goalie_height):
                attempts += 1
                # Brief flash to indicate save
                pygame.draw.circle(screen, WHITE, (int(puck_x), int(puck_y)), puck_radius*2)
                pygame.display.flip()
                pygame.time.delay(100)
                puck_active = False
                reset_puck()
            
            # Check if puck is out of bounds
            if puck_x > WIDTH:
                attempts += 1
                puck_active = False
                reset_puck()
        
        # Check for game over
        if attempts >= max_attempts and not game_won:
            game_active = False
            game_over = True
        
        # Draw player (shooter)
        screen.blit(player_img, (player_x, player_y))
        
        # Draw goalie
        screen.blit(goalie_img, (goalie_x, goalie_y))
        
        # Draw puck
        if puck_active or not game_active:
            screen.blit(puck_img, (int(puck_x - puck_radius - 1), int(puck_y - puck_radius - 1)))
        
        # Display controls at bottom
        controls_box = pygame.Rect(WIDTH//2 - 200, HEIGHT - 50, 400, 40)
        pygame.draw.rect(screen, LIGHTNING_BLUE, controls_box)
        pygame.draw.rect(screen, LIGHTNING_SILVER, controls_box, 2)
        
        controls_text = font.render("UP/DOWN: Move   SPACE: Shoot", True, WHITE)
        screen.blit(controls_text, (WIDTH//2 - controls_text.get_width()//2, HEIGHT - 40))
    
    # Game over screens
    if game_over:
        overlay = pygame.Surface((WIDTH, HEIGHT), pygame.SRCALPHA)
        overlay.fill((0, 0, 0, 180))
        screen.blit(overlay, (0, 0))
        
        # Game over box
        over_box = pygame.Rect(WIDTH//2 - 200, HEIGHT//2 - 100, 400, 200)
        pygame.draw.rect(screen, LIGHTNING_BLUE, over_box)
        pygame.draw.rect(screen, LIGHTNING_SILVER, over_box, 3)
        
        game_over_text = title_font.render("GAME OVER", True, WHITE)
        screen.blit(game_over_text, (WIDTH//2 - game_over_text.get_width()//2, HEIGHT//2 - 70))
        
        restart_text = font.render("Press R to Restart", True, WHITE)
        screen.blit(restart_text, (WIDTH//2 - restart_text.get_width()//2, HEIGHT//2))
        
        final_score = font.render(f"Final Score: {score}", True, WHITE)
        screen.blit(final_score, (WIDTH//2 - final_score.get_width()//2, HEIGHT//2 + 50))
    
    if game_won:
        overlay = pygame.Surface((WIDTH, HEIGHT), pygame.SRCALPHA)
        overlay.fill((0, 0, 0, 180))
        screen.blit(overlay, (0, 0))
        
        # Win box
        win_box = pygame.Rect(WIDTH//2 - 200, HEIGHT//2 - 100, 400, 200)
        pygame.draw.rect(screen, LIGHTNING_BLUE, win_box)
        pygame.draw.rect(screen, LIGHTNING_SILVER, win_box, 3)
        
        # Draw lightning bolts on both sides of the win message
        draw_lightning_bolt(screen, WIDTH//2 - 150, HEIGHT//2 - 70, 40)
        draw_lightning_bolt(screen, WIDTH//2 + 150, HEIGHT//2 - 70, 40)
        
        win_text = title_font.render("CHAMPION!", True, WHITE)
        screen.blit(win_text, (WIDTH//2 - win_text.get_width()//2, HEIGHT//2 - 70))
        
        champion_text = font.render("Tampa Bay Lightning Cup Winner!", True, LIGHT_BLUE)
        screen.blit(champion_text, (WIDTH//2 - champion_text.get_width()//2, HEIGHT//2))
        
        restart_text = font.render("Press R to Play Again", True, WHITE)
        screen.blit(restart_text, (WIDTH//2 - restart_text.get_width()//2, HEIGHT//2 + 50))
    
    # Update the display
    pygame.display.flip()
    
    # Cap the frame rate
    clock.tick(FPS)

# Quit the game
pygame.quit()
sys.exit()
```
